```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('NCI60_X (1).csv',index_col = 0)
```

```python
data
```

```python
data.corr()
```

```python
data.var()
```

```python
#KMeans Clustering before applying Principal Component Analysis
```

```python
from sklearn.cluster import KMeans
```

```python
inertia_list = []
```

```python
for i in range(1,11):
  km = KMeans(n_clusters = i,n_init = 20)
  km.fit(data)
  inertia = km.inertia_
  inertia_list.append(inertia)
```

```python
inertia_list
```

```python
plt.plot(range(1,11),inertia_list)
plt.xlabel("No of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
km = KMeans(n_clusters = 9, n_init = 20)
```

```python
km.fit(data)
```

```python
cluster_labels = km.predict(data)
```

```python
cluster_labels
```

```python
pd.Series(cluster_labels).value_counts()
```

```python
km.inertia_
```

```python
#Principal Component Analysis
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components = 30)
```

```python
pc_array = pc.fit_transform(data)
```

```python
pc_array.shape
```

```python
pc_data = pd.DataFrame(pc_array)
```

```python
pc_data
```

```python
pc_data.corr().round(1)
```

```python
pc.explained_variance_
```

```python
pc.explained_variance_ratio_ * 100
```

```python
#KMeans Clustering
```

```python
inertia_list = []
```

```python
for i in range(1,11):
  km = KMeans(n_clusters = i,n_init = 20)
  km.fit(pc_data)
  inertia = km.inertia_
  inertia_list.append(inertia)
```
