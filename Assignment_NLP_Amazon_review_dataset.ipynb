```python
full_data = pd.read_csv('/content/1429_1.csv')
```

```python
data = full_data[['reviews.text','reviews.title']]
data[['reviews.text','reviews.title']] = data[['reviews.text','reviews.title']].astype(str)
data.head()
```

```python
#LOWER CASING

data['reviews.text'] = data['reviews.text'].str.lower()
data['reviews.title'] = data['reviews.title'].str.lower()
data.head()
```

```python
#REMOVAL OF PUNCTUATION

def remove_punctuation(text):

    return text.translate(str.maketrans('', '',string.punctuation))

data['reviews.text'] = data['reviews.text'].apply(lambda text: remove_punctuation(text))
data['reviews.title'] = data['reviews.title'].apply(lambda text: remove_punctuation(text))
data.head()
```

```python
#REMOVING STOPWORDS

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):

    return " ".join([word for word in str(text).split() if word not in stop_words])

data['reviews.text'] = data['reviews.text'].apply(lambda text: remove_stopwords(text))
data['reviews.title'] = data['reviews.title'].apply(lambda text: remove_stopwords(text))
data.head()
```

```python
#REMOVAL OF FREQUENT WORDS

from collections import Counter
cnt = Counter()
for text in data['reviews.text'].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

```python
freq_words = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):

    return " ".join([word for word in str(text).split() if word not in freq_words])

data['reviews.text'] = data['reviews.text'].apply(lambda text: remove_freqwords(text))
```

```python
#REMOVAL OF RARE WORDS

n_rare_words = 10
rare_words = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
def remove_rarewords(text):

    return " ".join([word for word in str(text).split() if word not in rare_words])

data['reviews.text'] = data['reviews.text'].apply(lambda text: remove_rarewords(text))
```

```python
#REMOVAL OF FREQUENT WORDS

cnt = Counter()
for text in data['reviews.title'].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

```python
freq_words = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):

    return " ".join([word for word in str(text).split() if word not in freq_words])

data['reviews.title'] = data['reviews.title'].apply(lambda text: remove_freqwords(text))
```

```python
#REMOVAL OF RARE WORDS

n_rare_words = 10
rare_words = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
def remove_rarewords(text):

    return " ".join([word for word in str(text).split() if word not in rare_words])

data['reviews.title'] = data['reviews.title'].apply(lambda text: remove_rarewords(text))
```

```python
#LEMMATIZATION

import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

data['reviews.text'] = data['reviews.text'].apply(lambda text: lemmatize_words(text))
data['reviews.title'] = data['reviews.title'].apply(lambda text: lemmatize_words(text))
data.head()
```

```python
#STEMMING

from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

data['reviews.text'] = data['reviews.text'].apply(lambda text: stem_words(text))
data['reviews.title'] = data['reviews.title'].apply(lambda text: stem_words(text))
data.head()
```
